```python
d_path = sorted(os.listdir(os.getcwd()))[3] 
d_path = os.path.join(d_path, "logs")
d_path = os.path.join(d_path, os.listdir(d_path)[0])
d_path

all_loss = {"gen_loss": [], "disc_loss": []}
for summary in summary_iterator(d_path):
    try:
        tensor_bytes = summary.summary.value[0].tensor.tensor_content
        tensor_dtype = tf.dtypes.float32
        tensor_array = tf.io.decode_raw(tensor_bytes, tensor_dtype)
        valuee = np.float(tensor_array)
        loss_type=summary.summary.value[0].tag
        all_loss[loss_type].append(valuee)
    except Exception as e:
        print(e)
        
# a = 125
# df = pd.DataFrame(all_loss)
# print(df.shape)
# df = df.rolling(a).mean() 
# df = df.iloc[::4, :]

# df.plot()
```
